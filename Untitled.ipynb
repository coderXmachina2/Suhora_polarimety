```python
#Scrape data inventory

import glob
import re

x=glob.glob('./files_sorted/*')

c= 0
for a in x[:45]:
    b=glob.glob(a+'/*')
    for files in b:
        
        if 'pol_std' in files:
            print(files[len('./files_sorted\\'):-len('\\pol_std')])
            d = glob.glob(files+'/*')
            comp_array=[]
            for things in d:
                stringsearch = things[len('./files_sorted\\2020-03-05\\pol_std\\'):]

                result = re.search('(.*)-', stringsearch)
                comp_array.append(result.group(1)[:-2])
                
            print("Raw data polarization Standards:", 
                  ', '.join(list(set(comp_array)))) 
                  #', '.join(comp_array))
    c+=1
        
    print("\n")
```
